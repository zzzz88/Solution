```python
import pandas as pd
jeopardy  = pd.read_csv('jeopardy.csv')
```

```python
print(jeopardy.head(5))
jeopardy.columns = jeopardy.columns.str.replace(' ','')
jeopardy.columns
```

```python
jeopardy['clean_answer'] = jeopardy.Answer.str.lower().str.replace('\W',' ')
jeopardy['clean_question'] = jeopardy.Question.str.lower().str.replace('\W',' ')
```

```python
def value_normalize(value):
    clean_value = value.replace('$','')
    if clean_value.isdigit():
        return int(clean_value)
    else:
        return int(0)
jeopardy['clean_value'] = jeopardy.Value.apply(value_normalize)
```

```python
datetime = pd.to_datetime(jeopardy.AirDate)
datetime
```

```python
def match_counts(row):
    split_answer = row['clean_answer'].split()
    split_question = row['clean_question'].split()
    match_count = 0
    if 'the' in split_answer:
        split_answer.remove('the')
    if len(split_answer) == 0:
        return 0
    for value in split_answer:
        if value in split_question:
            match_count +=1
    return match_count/len(split_answer)
jeopardy['answer_in_question'] = jeopardy.apply(match_counts,
                                               axis =1)

        
    
```

```python
jeopardy.answer_in_question.mean()
```

```python

question_overlap = []
terms_used = set()

jeopardy = jeopardy.sort_values("AirDate",ascending = True)

for i, row in jeopardy.iterrows():
        split_question = row["clean_question"].split(" ")
        split_question = [q for q in split_question if len(q) > 5]
        match_count = 0
        for word in split_question:
            if word in terms_used:
                match_count += 1
        for word in split_question:
            terms_used.add(word)
        if len(split_question) > 0:
            match_count /= len(split_question)
        question_overlap.append(match_count)
jeopardy["question_overlap"] = question_overlap

jeopardy["question_overlap"].mean()
```

```python
def separate_values(row):
    if row.clean_value > 800:
        value = 1
    else:
        value = 0
    return value
jeopardy['high_value'] = jeopardy.apply(separate_values, axis =1)
def count(word):
    low_count=0
    high_count = 0
    for index, row in jeopardy.iterrows():
        clean_question = row.clean_question.split(" ")
        if word in clean_question:
            if row.high_value == 1:
                high_count +=1
            else:
                low_count +=1
    return high_count, low_count
observed_expected = list()
comparison_terms = list(terms_used)[:5]
for item in comparison_terms:
    high_low_count =count(item)
    observed_expected.append(high_low_count)
print(observed_expected)
```

```python
from scipy.stats import chisquare
high_value_count = (jeopardy.high_value==1).sum()
low_value_count = (jeopardy.high_value==0).sum()
chi_squared =list()
for item in observed_expected:
    total = item.sum()
    total_prop = total/jeopardy.shape[0]
    expected_high_rows = total_prop*high_value_count
    expected_low_rows = total_prop*low_value_count
    observed = [item[0],item[1]]
    expected = [expected_high_rows,expected_low_rows]
    result = chisquare(observed,expected)
    chi_squared.append(result)
print(chi_squared)
```
